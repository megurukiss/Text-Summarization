```python
import matplotlib.pyplot as plt
import pandas as pd
baseline=pd.read_csv('./results/personalized_summary_results_exp1.csv')
prompt_with_examples=pd.read_csv('./results/personalized_summary_results_bootstrap_exp2.csv')
prompt_refined=pd.read_csv('./results/personalized_summary_results_miprov_exp3.csv')

def plot_score_comparison():
    """
    Create a simple line plot comparing summary_scores and gt_scores.
    
    Parameters:
    df (pandas.DataFrame): DataFrame containing 'summary_scores' and 'gt_scores' columns
    """
    # Create figure and axis
    plt.figure(figsize=(10, 6))
    
    # Create line plots for both scores
    plt.plot(baseline['summary_scores'], marker='o', label='Summary Scores')
    plt.plot(baseline['gt_scores'], marker='o', label='baseline prompt')
    plt.plot(prompt_with_examples['summary_scores'], marker='o', label='refined prompt1')
    plt.plot(prompt_refined['summary_scores'], marker='o', label='refined prompt2')
    
    # Add labels and title
    plt.xlabel('Document Index')
    plt.ylabel('Scores')
    plt.title('Summary Scores vs Ground Truth Scores')
    
    # Add legend
    plt.legend()
    
    # Add grid for better readability
    plt.grid(True, alpha=0.3)
    
    # Adjust layout
    plt.tight_layout()
    
    plt.show()
plot_score_comparison()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_score_comparisons():
    """
    Create three subplots comparing summary_scores and gt_scores for each experiment.
    """
    # Create figure and axes for subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))
    
    # Plot for baseline experiment
    ax1.plot(baseline['summary_scores'], marker='o', label='Summary Scores', color='blue')
    ax1.plot(baseline['gt_scores'], marker='o', label='Ground Truth Scores', color='red')
    ax1.set_title('Baseline Experiment')
    ax1.set_xlabel('Document Index')
    ax1.set_ylabel('Scores')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot for prompt with examples experiment
    ax2.plot(prompt_with_examples['summary_scores'], marker='o', label='Summary Scores', color='blue')
    ax2.plot(baseline['gt_scores'], marker='o', label='Ground Truth Scores', color='red')
    ax2.set_title('Prompt with Examples Experiment')
    ax2.set_xlabel('Document Index')
    ax2.set_ylabel('Scores')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot for refined prompt experiment
    ax3.plot(prompt_refined['summary_scores'], marker='o', label='Summary Scores', color='blue')
    ax3.plot(baseline['gt_scores'], marker='o', label='Ground Truth Scores', color='red')
    ax3.set_title('Refined Prompt Experiment')
    ax3.set_xlabel('Document Index')
    ax3.set_ylabel('Scores')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    
    plt.show()

# Call the function
plot_score_comparisons()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_score_boxplots():
    """
    Create a box plot comparing summary scores and ground truth scores across all experiments.
    """
    plt.figure(figsize=(12, 6))
    
    # Prepare data for box plot
    data = [
        baseline['summary_scores'],
        baseline['gt_scores'],
        prompt_with_examples['summary_scores'],
        prompt_refined['summary_scores'],
    ]
    
    # Create labels for each box
    labels = [
        'Baseline\nSummary',
        'Baseline\nGround Truth',
        'Examples\nSummary',
        'Refined\nSummary',
    ]
    
    # Create box plot
    box_plot = plt.boxplot(data, labels=labels, patch_artist=True)
    
    # Customize box colors
    colors = ['lightblue', 'lightcoral'] * 3  # Alternate colors for summary and ground truth
    for patch, color in zip(box_plot['boxes'], colors):
        patch.set_facecolor(color)
    
    # Customize plot
    plt.title('Score Distribution Comparison Across Experiments')
    plt.ylabel('Scores')
    plt.grid(True, axis='y', alpha=0.3)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=0)
    
    # Add legend
    plt.axhline(y=-1, color='lightblue', label='Summary Scores', xmin=0.95)
    plt.axhline(y=-1, color='lightcoral', label='Ground Truth Scores', xmin=0.95)
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Call the function
plot_score_boxplots()
```
