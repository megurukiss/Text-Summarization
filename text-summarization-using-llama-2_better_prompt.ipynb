```python
from transformers import AutoTokenizer
import transformers
import torch
import warnings
warnings.filterwarnings('ignore')
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
import pandas as pd

data=pd.read_csv('/root/Text-Summarization/summarizations_with_ratings.csv')
```

```python
data
```

```python
article=data['article']
```

```python
tokenizer = AutoTokenizer.from_pretrained("togethercomputer/LLaMA-2-7B-32K")
model = AutoModelForCausalLM.from_pretrained("togethercomputer/LLaMA-2-7B-32K")
model = model.to("cuda")
```

```python
article=article.to_list()
```

```python
summarizations=[]
for i in article:
    input_context = "Your task is to generate a formal, concise, 50-word summary that clearly highlights the key points and main ideas of the document: " + i
    # limit the length of input_context to 1024
    input_context = input_context[:1536]
    input_ids = tokenizer.encode(input_context, return_tensors="pt").to("cuda") 
    output = model.generate(input_ids, max_new_tokens=128, temperature=0.7)
    output_text = tokenizer.decode(output[0], skip_special_tokens=True)

    # split output_text to only have Answer, plit from '\nA: '
    answer= output_text[1536:]
    summarizations.append(answer.replace("\n", ""))
    print('anwer:',answer)

print('len of summarizations:',len(summarizations))
    
```

```python
data['summarizations_prompt']=summarizations
```
